### ***1 Mengubah Tipe Data dan Memperbaiki Kejanggalan Kolom date_received & incident_date***

Pada bagian ini kita akan mengubah tipe data untuk date_recieved dan incident_date dari object menjadi datetime. Dan juga memperbaiki kejanggalan yang terjadi pada kedua kolom tersebut dimana tanggal laporan lebih awal dari tanggal kejadian. Untuk permasalahan tersebut kita akan lakukan swap data.

```python
# mengubah tipe data kolom tanggal
df['date_received'] = pd.to_datetime(df['date_received'], errors='coerce')
df['incident_date'] = pd.to_datetime(df['incident_date'], errors='coerce')
df.dropna(subset=['date_received', 'incident_date'], inplace=True)
```

```python
# menghapus data yang tidak diperlukan untuk memastikan rentang tahun dari 2002 s/d 2015
df.drop(df[df['date_received'].dt.year < 2002].index, inplace=True)
df.drop(df[df['incident_date'].dt.year < 2002].index, inplace=True)
df.drop(df[df['date_received'].dt.year > 2015].index, inplace=True)
df.drop(df[df['incident_date'].dt.year > 2015].index, inplace=True)
```

```python
df.info()
df.head()
```

```python
#incident_date dan date_received menjadi timestamp
df['incident_date'] = df['incident_date'].values.astype(np.int64) // 10 ** 9
df['date_received'] = df['date_received'].values.astype(np.int64) // 10 ** 9
df.head()
```

```python
# memindahkan tanggal yang lebih awal di kolom date_received ke incident_date, sebagai perbaikan kejanggalan yang terjadi di kedua kolom tersebut.
situation = df.incident_date > df.date_received
df.loc[situation, ['incident_date', 'date_received']] = df.loc[situation, ['date_received', 'incident_date']].values
df['incident_date'] = pd.to_datetime(df['incident_date'], unit='s')
df['date_received'] = pd.to_datetime(df['date_received'], unit='s')
df.head()
```

### ***2 Merapikan kolom status***


Pada proses ini, beberapa data unik yang memiliki maksud yang sama di satukan dengan 4 value inti, yaitu
- Approved (klaim disetujui dan diganti penuh (Approve In Full))
- Denied (klaim yang diajukan ditolak)
- Settled (klaim yang diajukan disetujui namun nominal yang diberikan hanya sebagian dari klaim yang diajukan)
- Canceled (klaim yang diajukan oleh penumpang dibatalkan oleh penumpang atau pihak lain)


```python
# menghitung jumlah data pada tiap data unik di kolom status
df['status'].value_counts()
```

```python
# mengubah beberapa data unik kedalam 4 kategori data unik utama
df.loc[(df['status'].isin(['Approve in Full'])),'status']='Approved'
df.loc[(df['status'].isin(['Deny'])),'status']='Denied'
df.loc[(df['status'].isin(['Settle'])),'status']='Settled'
df.loc[(df['status'].isin(['Insufficient; one of the following items required: sum certain; statement of fact; signature; location of incident; and date.'])),'status']='Denied'

df['status'].value_counts()
```

```python
# Menghapus data unik yang tidak bisa masuk ke dalam 4 kategori utama
df.drop(df[df['status'] == 'In review'].index, inplace=True)
df.drop(df[df['status'] == 'In litigation'].index, inplace=True)
df.drop(df[df['status'] == 'Claim has been assigned for further investigation'].index, inplace=True)
df.drop(df[df['status'] == 'Pending response from claimant'].index, inplace=True)
df.drop(df[df['status'] == 'Closed as a contractor claim'].index, inplace=True)
df.drop(df[df['status'] == '-'].index, inplace=True)

display(df['status'].isna().sum())
display(df['status'].value_counts())
```

### ***3. Mengatur kolom disposition***

Kolom disposition ini berhubungan dengan kolom status yang sudah kita rapikan sebelumnya. Setelah kita mengisi dengan benar kolom status maka kolom disposition juga bisa kita isi. 

```python
# cek jumlah data kosong di kolom disposition sesuai dengan value di kolom status
df[df['disposition'].isna()]['status'].value_counts()
```

***Proses selanjutnya kita akan mengisi data kosong di kolom `disposition` yang berisi 'Approved' di kolom `status` dengan 'Approve in full', kolom `disposition` yang berisi 'Settled' di kolom `status` dengan 'Settle', kolom `disposition` yang berisi 'Canceled' di kolom `status` dengan 'Cencel', kolom `disposition` yang berisi 'Denied' di kolom `status` dengan 'Deny'***

```python
df.loc[(df['disposition'].isna())&(df['status'].isin(['Approved'])),'disposition']='Approve in Full'

df.loc[(df['disposition'].isna())&(df['status'].isin(['Settled'])),'disposition']='Settle'

df.loc[(df['disposition'].isna())&(df['status'].isin(['Canceled'])),'disposition']='Cancel'

df.loc[(df['disposition'].isna())&(df['status'].isin(['Denied'])),'disposition']='Deny'

display(df['disposition'].isna().sum())
display(df['disposition'].value_counts())
```

### ***4 Mengubah Tipe Data claim_amount & close_amount ke tipe float***

Sudah seharusnya tipe data dari claim_amount & close_amount merupakan tipe data angka karena kolom tersebut adalah kolom nominal uang. Selanjutnya kita akan mengubah tipe data ke float. Pengubahan data ini dapat mempermudah analisa kita pada proses selanjutnya.

```python
# mengetahui jumlah value yang kosong dikolom claim_amount
df['claim_amount'].isna().sum()
```

```python
# menghapus seluruh row yang memiliki value kosong dikolom claim_amount
df.dropna(subset=['claim_amount'], inplace = True)
df['claim_amount'].isna().sum()
```

```python
# menghapus karakter-karakter pada kolom close_amount dan claim_amount yang tidak relevant
df['close_amount'] = df['close_amount'].astype('str')
df['close_amount'] = df['close_amount'].str.replace('$','', regex=False)
df['close_amount'] = df['close_amount'].str.replace(',', '',regex=False)
df['close_amount'] = df['close_amount'].str.replace(';', '',regex=False)
df['close_amount'] = df['close_amount'].astype('float')

df['claim_amount'] = df['claim_amount'].astype('str')
df['claim_amount'] = df['claim_amount'].str.replace('$', '', regex=False)
df['claim_amount'] = df['claim_amount'].str.replace(',', '',regex=False)
df['claim_amount'] = df['claim_amount'].str.replace(';', '',regex=False)
df['claim_amount'] = df['claim_amount'].str.replace('-', '',regex=False)
df['claim_amount'] = df['claim_amount'].astype('float')

display(df.info(),df.head())
```

```python
# Menghapus data yang memiliki claim_amount diatas 200.000 dolar karena terlalu jauh persebaran datanya
df.drop(df.loc[df['claim_amount']>200000].index, inplace=True)
```

### ***5. Mengorganisir kolom close_amount***


Untuk mengisi nilai pada kolom close_amount, kita dapat mengecek nilai pada kolom status terlebih dahulu. Jika nilai pada kolom status bernilai Approved, maka kolom close_amount dapat diisi dengan nilai yang sama dengan nilai pada kolom claim_amount. Sebaliknya, jika nilai pada kolom status bernilai Settled, maka kolom close_amount dapat diisi dengan setengah dari nilai pada kolom claim_amount. Dengan demikian, dengan mengetahui nilai pada kolom status, kita dapat mengisi nilai pada kolom close_amount dengan benar.

```python
# melihat jumlah data kosong pada kolom claim_amount yang sesuai dengan nilai pada kolom status
df[df['close_amount'].isna()]['status'].value_counts()
```

***Proses selanjutnya adalah mengisi data kosong di kolom `disposition` yang berisi 'Canceled' di kolom `status` dengan 'Cencel', kolom `disposition` yang berisi 'Denied' di kolom `status` dengan 'Deny', di kolom `close_amount` yang berisi 'Approved' di kolom `status` dengan nilai yang sama dengan kolom 'claim_amount', di kolom `close_amount` yang berisi 'Settled' di kolom `status` dengan nilai setengah dari nilai 'claim_amount'***

```python
df.loc[(df['close_amount'].isna())&(df['status'].isin(['Canceled'])),'close_amount']= 0

df.loc[(df['close_amount'].isna())&(df['status'].isin(['Denied'])),'close_amount']= 0

df.loc[(df['close_amount'].isna())&(df['status'].isin(['Approved'])),'close_amount']= df['claim_amount']

df.loc[(df['close_amount'].isna())&(df['status'].isin(['Settled'])),'close_amount']= df['claim_amount']/2

display(df['close_amount'].isna().sum())
df.tail()
```

### **Analisis Menggunakan Statistik dan Melihat Korelasi (analisa tambahan)** 

```python
# Membuat tabel baru yang hanya berisi klaim dengan status Settled dan Approved
df_settle = df.loc[(df['status'] == 'Settled')]
df_full = df.loc[(df['status'] == 'Approved')]
```

```python
#Untuk membuat kolom baru cat_nominal yang mengkategorikan nominal besar (1) dan kecil (0) berdasarkan batas pengajuan klaim sebesar 5000 dollar
#Serta kolom baru hasil yang mengindikasikan apakah nominal yang disetujui (settle) melebihi 50% dari nominal pengajuan (claim_amount),def cat_nominal(i)

def cat_nominal(i):
    if i < 5000:
        return 0
    else:
        return 1
 
df_settle['cat_nominal'] = df_settle['close_amount'].map(cat_nominal)

df_settle.loc[df_settle['close_amount'] <= (df_settle['claim_amount']/2), 'hasil'] = 'False'
df_settle.loc[df_settle['close_amount'] > (df_settle['claim_amount']/2), 'hasil'] = 'True'
df_settle.head()

# df['cat_nominal'] = np.where(df['claim_amount'] > 5000, 1, 0)
# df['hasil'] = np.where(df['settle_amount'] / df['claim_amount'] > 0.5, 'settle', 'not settle')
```

```python
# cek distribusi kolom cat_nominal
from scipy.stats import normaltest
stats, pval=normaltest(df_settle['cat_nominal'])
if pval<=0.05:
    print('tidak normal') #Ha
else: 
    print('distribusi normal') #Ho 
```

#### *Untuk menguji apakah kategori nominal yang diterima penumpang (kolom cat_nominal) mempengaruhi total nominal yang diterima oleh penumpang, Anda dapat menggunakan uji statistik Mann-Whitney. Uji ini digunakan untuk membandingkan dua kelompok data yang distribusinya tidak normal.*

```python
from scipy.stats import mannwhitneyu
stats,pval=mannwhitneyu(*[df_settle[df_settle['hasil']==i]['cat_nominal'] for i in df_settle['hasil'].unique()], alternative='less')
if pval <= 0.05:
    print('Tolak Ho')
else:
    print('Gagal menolak Ho')
```

```python
# Menambah kolom baru interval
df['interval'] = (df['date_received'] - df['incident_date']).dt.days
df.head()
```

```python

# Membuat histogram interval pengajuan
sns.distplot(df['interval'])
plt.title('Interval histogram', size = 15)
plt.show()
```

```python
# cek distribusi kolom interval
from scipy.stats import normaltest
stats, pval=normaltest(df['interval'])
if pval<=0.05:
    print('tidak normal') #Ha
else: 
    print('distribusi normal') #Ho 
```

#### *Untuk menguji apakah interval tanggal insiden dan tanggal laporan diterima mempengaruhi status klaim, Anda dapat menggunakan uji statistik Kruskal-Wallis. Uji ini digunakan untuk membandingkan lebih dari dua kelompok data yang distribusinya tidak normal.*

```python
from scipy.stats import kruskal
stats,pval=kruskal(*[df[df['status']==i]['interval'] for i in df['status'].unique()])
if pval <= 0.05:
    print('Tolak Ho')
else:
    print('Gagal menolak Ho')
```
